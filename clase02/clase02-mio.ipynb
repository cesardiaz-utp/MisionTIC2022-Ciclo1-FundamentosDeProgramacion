# Listas [ ]

```python
# las lista se pueden modificar
listaInt = [1,2,3]
print(listaInt)
print(listaInt[0])
listaStr = ["a","b",'c',[4,5,6]]
print(listaStr)
print(listaStr[2])
print(listaStr[3])
print(listaStr[3][1])
print(listaStr[3][1] + listaStr[3][2])
```

# Tuplas ( )

```python
# las tuplas después de creadas no las puedo cambiar INMUTABLES
tuplaInt = (1,2,3,4)
print(tuplaInt)
print(tuplaInt[1])

tuplaStr = ("A", "B", "C")
print(tuplaStr[0])

tuplasMixtas = (1.0, 2, "3", [1, 2, 3])
print(tuplasMixtas)
```

# Rangos

```python
r1 = range(7)



print(r1)
print(r1[0])
print(r1[6])
# print(r1[7]) ERROR genera
for i in r1:
    print("r1",i)



r2 = range(1,10)
print(r2)
print(r2[0])
print(r2[8])
# print(r2[9]) ERROR genera
for i in r2:
    print("r2",i)

r3 = range(0, 10, 2)
print(r3)
print(r3[0])
print(r3[4])
# print(r3[5]) ERROR genera
for i in r3:
    print("r3",i)
```
