# Precedencia en operadores aritméticas

1. Paréntesis()
2. Exponenciacion **
3. Negación de expresiones -
4. Multiplicación *, Division /, Division entera //, Modulo %
5. Adición +, Sustracción -
6. Asignación =  

```python
precedencia = 13 - 4 * -(5-2) + 3 * ( 2+8)
print('pre: ', precedencia)

precedencia_1 = 13 - 4 * -(+3) + 3 * 10 
print('pre 1 : ', precedencia_1)

precedencia_2 = 13 - 4 * -3 + 3 * 10 # resuelve 3. Negación de expresiones -
print('pre 2 : ', precedencia_2)

precedencia_3 = 13 - -12 + 30 # resuelve 4. Multiplicación *, Division /, Division entera //, Modulo %
print('pre 3 : ', precedencia_3)

precedencia_4 = 13  + 12 + 30 # resuelve 3. Negación de expresiones -
print('pre 4 : ', precedencia_4) # 5. Adición +, Sustracción -
```
